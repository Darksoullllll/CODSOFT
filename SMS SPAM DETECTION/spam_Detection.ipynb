```python
import pandas as pd
import numpy as np
import plotly.express as px
df = pd.read_csv("spam.csv",encoding='latin1')
df.head()
```

DATA PREPROCESSING

```python
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis='columns',inplace=True)
```

```python
df
```

```python
df.rename({'v1':'spam_or_not','v2':'language'},axis='columns',inplace=True)
```

```python
df
```

Encoding Categorical value

```python
from sklearn.preprocessing import LabelEncoder
le =  LabelEncoder()
df['spam_or_not_encoded'] = le.fit_transform(df['spam_or_not'])
```

```python
df
```

Deleting the Spam or not column

```python
df.drop(['spam_or_not'],axis='columns',inplace=True)
```

```python
df
```

Data Visualization

```python
spam = (df['spam_or_not_encoded'] == 1).sum()
ham = (df['spam_or_not_encoded'] == 0).sum()
fig = px.bar(x = ['spam = 1','ham = 0'],y = [spam,ham],title="Spam and Ham Mails",)
fig.update_xaxes(title_text="Spam & Ham")
fig.update_yaxes(title_text="Counts")
```

Data set contains the ham mail in excess amount 

```python
df['lan_len'] = df['language'].apply(len)
fig = px.histogram(df,'lan_len',color='spam_or_not_encoded')
fig.show()
```

From the histogram the ham mail have typical length of 0 -100
Whereas the spam have the length between 120 - 170


Cleaning the message 

```python
df.drop(['lan_len'],axis='columns',inplace=True)
df.tail()
```

```python
import re
import nltk  #help in removing stocks words i.e and ,the,a,or
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer #change different word into one word which have same meaning
corpus = [] #contains the cleaned data
for i in range(0,5572):
    review=re.sub('[^a-zA-Z]', ' ', df['language'][i])#remove special characters from text
    review=review.lower()   #convert to lower case
    review=review.split()
    ps = PorterStemmer()
    all_stop = stopwords.words('english')
    all_stop.remove('not')
    review = [ps.stem(word) for word in review if not word in set(all_stop)] 
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

Creating Bag of Words to make message to machine understandable language

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x= cv.fit_transform(corpus).toarray()
y= df.iloc[:,-1].values

import joblib
joblib.dump(cv, 'count_vectorizer.pkl')
```

```python
x
```

```python
y
```

Splitting the data into train and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
y_train
```

Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y_pred = gnb.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
mat1 = confusion_matrix(y_test,y_pred)
score1 = accuracy_score(y_test,y_pred)
rep1 = classification_report(y_test,y_pred)
print("Confusion Matrix: ")
print(mat1)
print()
print("Accuracy Score: ",score1 * 100)
print()
print("Report: ")
print(rep1)
```

Mulitple Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
mnb_pred = mnb.predict(x_test)
mnb_pred
```

```python
mat2 = confusion_matrix(y_test,mnb_pred)
score2 = accuracy_score(y_test,mnb_pred)
rep2 = classification_report(y_test,mnb_pred)
print("Confusion Matrix: ")
print(mat2)
print()
print("Accuracy Score: ",score2 * 100)
print()
print("Report: ")
print(rep2)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr_pred = lr.predict(x_test)
lr_pred
```

```python
mat3 = confusion_matrix(y_test,lr_pred)
score3 = accuracy_score(y_test,lr_pred)
rep3 = classification_report(y_test,lr_pred)
print("Confusion Matrix: ")
print(mat3)
print()
print("Accuracy Score: ",score3 * 100)
print()
print("Report: ")
print(rep3)
```

SVM

```python
from sklearn.svm import SVC
svm =SVC(kernel='rbf')
svm.fit(x_train,y_train)
```

```python
svm_pred = svm.predict(x_test)
```

```python
mat4 = confusion_matrix(y_test,svm_pred)
score4 = accuracy_score(y_test,svm_pred)
rep4 = classification_report(y_test,svm_pred)
print("Confusion Matrix: ")
print(mat4)
print()
print("Accuracy Score: ",score4 * 100)
print()
print("Report: ")
print(rep4)
```
